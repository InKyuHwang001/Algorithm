### 재귀함수

```python
# 더하기
def f(x):
  if x<=1:
    1
  else:
    x+f(x-1)
# 곱하기
def f(x):
  if x<=1:
    1
  else:
    x*f(x-1)
```

## 2진수 구하기

```python
x= int(input('2진수로 바꿀 수를 입력하여라:'))
result=''
while True:
  if x%2==0:
    result+='0'
  else:
    result+='1'
  x=x//2
  if x==1 or x==0:
    result+=str(x)
    print(result[::-1])
    break
#한번더 보아요~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
```

```python
def dinmaker(x):
  if x<2:
    return str(x)
  else:
    return str(dinmaker(x//2))+str(x%2)
print(dinmaker(11))
```

## 자릿수 더하기

```python
x=0
for i in '2231':
  x+=int(i)
print(x)
```

```python
def adding(x):
  if x=="":
    return 0
  else: 
    return int(x[0])+int(adding(x[1:]))
print(adding('2231'))
```

```python
def adding(x):
  if x<10:
    return x
  else:
    return adding(x//10)+(x%10)
adding(11)
```

## 숫자 세기

```python
def numbercount(number):
  if number<=0:
    print('끝')
  else:
    print(number)
    return numbercount(number-1)
numbercount(10)
```

## 피보나치

```python
def pivo(x):
  if x==0 or x==1:
    return 1
  return pivo(x-1)+pivo(x-2)
print(pivo(5))
```

```python
def factory(x):
  if x==0 or x==1:
    return 1
  else:
    return x*factory(x-1)
factory(10)
```

### 리스트 더하기

```python
s=0
for i in [100,200,300,400]:
  s+=i
print(s)
```

```python
def sum(x):
  if len(x)==1:
    return x[0]
  else:
    return x[0]+sum(x[1:])
x=[100,200,300,400]
sum(x)
```

### 승수

```python
print(2**6)
result=1
for _ in range(6):
  result*=2
print(result)
```

```python
def mult(n,e):
  result=1
  for _ in range(e):
    result*=2
  print(result)
mult(2,6)
```

```python
def mult(n,e):
  if e==1:
    return n
  else:
    return n*mult(n,e-1)
print(mult(2,6))
```
