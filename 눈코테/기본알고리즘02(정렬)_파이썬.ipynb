## 선택정렬

```python
a=[5,10,66,77,54,32,11,15]
sorted=[]

while a:
  sorted.append(min(a))
  a.pop(a.index(min(a)))
  print(sorted)
  print(a)
  print('---------')
```

```python
a=[5,10,66,77,54,32,11,15]

def minval(l):
  min_=l[0]
  for i in l:
    if min_>i:
      min_=i
  return min_

minval(a)
```

```python
a=[1,50,10,66,77,54,32,11,15]

def minindex(l):
  min_=l[0]
  count=0
  for i in l:
    if min_>i:
      min_=i
      count+=1
  return count

print(minindex(a))
a=[1,50,10,66,77,54,32,11,15]
```
